```python
root_path = r'k:\dropbox\maps\garmin etrex 30\2013'
fns = []
for root, dirs, files in os.walk(root_path):
    gpxfns = glob.glob(os.path.join(root, '*.gpx'))
    if gpxfns:
        print '% 9.0f %s' % (len(gpxfns), root)
    fns += gpxfns
print '---------'
print '% 9.0f Total' % len(fns)
```

```python
import pyxie.io
```

```python
darr = np.empty((0, 4), dtype=float)
dsets = {}
for fn in fns:
    arr = pyxie.io.read_gpx(fn)
    i_from = darr.shape[0]
    darr = np.row_stack((darr, arr))
    i_to = darr.shape[0]
    dsets[fn] = (i_from, i_to)
    sorted_args = np.argsort(darr[:,0])
    if len(sorted_args) > 0:
        darr = darr[sorted_args]
```

The below filters out those really annoying jumps at the start of a GPX file.

```python
def distance(pt1, pt2):
    return np.sqrt((pt2[0]-pt1[0])**2 + (pt2[1]-pt1[1])**2)

nfig = 0
for i in range(len(arrs)):
    arr = arrs[i]
    
    dist = np.zeros(arr.shape[0])
    for j in range(len(arr)):
        if j == 0:
            dist[j] = distance(arr[j+1,1:3], arr[j,1:3])
        elif j == len(arr)-1:
            dist[j] = distance(arr[j-1,1:3], arr[j,1:3])
        else:
            dist[j] = np.mean((distance(arr[j+1,1:3], arr[j,1:3]), distance(arr[j,1:3], arr[j-1,1:3])))
    
    # A distance of how many standard deviations is a 'bad' jump?
    max_std = np.std(dist) * 6
    flag = False
    deleted = []
    for j in range(len(dist)):
        if len(dist) > 0:
            if dist[j] > max_std:
                deleted.append(list(arr[j,1:3]))
                arrs[i][j] = np.empty(4) * np.nan
                flag = True
    if flag:
        deleted = np.asarray(deleted)
        print i, deleted
        if nfig and 0:
            fig = plt.figure()
            ax = fig.add_subplot(111, aspect=True)
            ax.plot(arr[:,1], arr[:,2])
            ax.scatter(deleted[:,0], deleted[:,1])
            nfig += 1
```

```python
fig = plt.figure()
gs = GridSpec(1, 1)
ax = fig.add_subplot(gs[0], aspect=True)
for arr in arrs:
    ax.plot(arr[:,1], arr[:,2])
ax.set_title('after cleaning')
```
