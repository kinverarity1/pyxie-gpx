The below filters out those really annoying jumps at the start of a GPX file.

```python
def distance(pt1, pt2):
    return np.sqrt((pt2[0]-pt1[0])**2 + (pt2[1]-pt1[1])**2)

nfig = 0
for i in range(len(arrs)):
    arr = arrs[i]
    
    dist = np.zeros(arr.shape[0])
    for j in range(len(arr)):
        if j == 0:
            dist[j] = distance(arr[j+1,1:3], arr[j,1:3])
        elif j == len(arr)-1:
            dist[j] = distance(arr[j-1,1:3], arr[j,1:3])
        else:
            dist[j] = np.mean((distance(arr[j+1,1:3], arr[j,1:3]), distance(arr[j,1:3], arr[j-1,1:3])))
    
    # A distance of how many standard deviations is a 'bad' jump?
    max_std = np.std(dist) * 6
    flag = False
    deleted = []
    for j in range(len(dist)):
        if len(dist) > 0:
            if dist[j] > max_std:
                deleted.append(list(arr[j,1:3]))
                arrs[i][j] = np.empty(4) * np.nan
                flag = True
    if flag:
        deleted = np.asarray(deleted)
        print i, deleted
        if nfig and 0:
            fig = plt.figure()
            ax = fig.add_subplot(111, aspect=True)
            ax.plot(arr[:,1], arr[:,2])
            ax.scatter(deleted[:,0], deleted[:,1])
            nfig += 1
```
